In this work, effect of high dimensionality and big sample size to clustering and outlier detecting will be demonstrated through some examples. Some image manipulations will be done to understand the dimensionality and structure of rgb images.

```python
#Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import math
from scipy.spatial import distance
import matplotlib.image as mpimg
```

A function is created to generate a dataframe which containt n dimensional points in desired size. Another function is created to calcualate euclidean distance to origin of one point.

```python
#setting seed
random.seed(100)
def Dimension_generator(n,size):
    Dimension = pd.DataFrame()
    for i in list(range(0,n)):
        Dimension[i] = np.random.uniform(low=-1.0, high=1.0, size=size)
    return Dimension

def Distance_to_origin(series):
    return sum(series**2)**.5
```

Creating 1000 random points distributed uniformly from -1 to 1 in 15 dimensions

```python
Dimensions = np.empty(shape = 15,dtype = object)
for i in list(range(0,15)):
    Dimensions[i] = Dimension_generator(i+1,1000)
```

Calculating each point's distance to origin

```python
distance_to_origin = pd.DataFrame()
for i in list(range(0,15)):
    for j in list(range(0,1000)):
        distance_to_origin.loc[j,i] = Distance_to_origin(Dimensions[i].loc[j])
```

Calculating all fractions of point withing 1 distance to origin

```python
fractions = np.empty(15,dtype = object)
for i in list(range(0,15)):
    fractions[i] = sum(distance_to_origin[i]<1)/1000
```

```python
fractions
```

As the dimensionality increases, points are going further away from origin. Here we can define origin as cluster center. So, it basicly makes the clustering and outlier detecting harder. Because every points are further away from eacher and cluster center than in low dimensionality. Because points are far away from each other, it is harder to find outliers.

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(list(range(1,16)),
          fractions,
          color = "black")
ax.set(title= "Fractions versus Dimension",
      xlabel = "Dimensions",
      ylabel = "Fractions")

plt.show()
```

For 2D fraction = area of circle to square inscribing it = pir^2/4r^2
so for 2D case pi estimation = 4*fraction
for 3D fraction = volume of circle to square inscribing it = 4/3pir^3/8r^3
so for 3D case pi estimation = 6*fraction

```python
epi2d = 4*fractions[1]
epi3d = 6*fractions[2]
print(epi2d)
print(epi3d)
```

```python
def epi_estimator2d(size):
    D2 = np.empty(shape = 1,dtype = object)
    D2[0] = Dimension_generator(2,size)
    distance_to_origin2d = pd.DataFrame()
    for j in list(range(0,size)):
        distance_to_origin2d.loc[j,0] = Distance_to_origin(D2[0].loc[j])
    fractions2d = sum(distance_to_origin2d[0]<1)/size
    epi2d = 4*fractions2d
    return epi2d

def epi_estimator3d(size):
    D3 = np.empty(shape = 1,dtype = object)
    D3[0] = Dimension_generator(3,size)
    distance_to_origin3d = pd.DataFrame()
    for j in list(range(0,size)):
        distance_to_origin3d.loc[j,0] = Distance_to_origin(D3[0].loc[j])
    fractions3d = sum(distance_to_origin3d[0]<1)/size
    epi3d = 6*fractions3d
    return epi3d
```

```python
estimatedpis2d = np.empty(shape = 6,dtype = object)
estimatedpis2d[0] = epi2d
estimatedpis2d[1] = epi_estimator2d(5000)
estimatedpis2d[2] = epi_estimator2d(10000)
estimatedpis2d[3] = epi_estimator2d(25000)
estimatedpis2d[4] = epi_estimator2d(50000)
estimatedpis2d[5] = epi_estimator2d(100000)

estimatedpis3d = np.empty(shape = 6,dtype = object)
estimatedpis3d[0] = epi3d
estimatedpis3d[1] = epi_estimator3d(5000)
estimatedpis3d[2] = epi_estimator3d(10000)
estimatedpis3d[3] = epi_estimator3d(25000)
estimatedpis3d[4] = epi_estimator3d(50000)
estimatedpis3d[5] = epi_estimator3d(100000)
```

Estimated pi's are created for different sample sizes. We can observe that as sample size gets bigger variability decreases and estimates pi getting closer to real value of pi. Getting closer to real pi is harder in 3D than 2D.

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter([1000,5000,10000,25000,50000,100000],
          estimatedpis2d,
          color = "yellow")
ax.scatter([1000,5000,10000,25000,50000,100000],
          estimatedpis3d,
          color = "blue")
plt.plot([0,100000],[math.pi,math.pi],
       color = "darkgreen")
ax.set(title= "PI Estimation",
      xlabel = "Sample Size",
      ylabel = "Esimated PI")

plt.show()
```

```python
newDimensions = np.empty(shape = 15,dtype = object)
for i in list(range(0,15)):
    newDimensions[i] = Dimension_generator(i+1,1000)

testDimensions = np.empty(shape = 15,dtype = object)
for i in list(range(0,15)):
    testDimensions[i] = Dimension_generator(i+1,100)
```

```python
def mindistance(test,new):
    mindist = distance.euclidean(test,new.loc[0])
    for i in list(range(0,999)):
        if distance.euclidean(test,new.loc[i]) < mindist:
            mindist = distance.euclidean(test,new.loc[i])
    return mindist    
```

```python
mindistances = pd.DataFrame()
for i in list(range(0,15)):
    for j in list(range(0,100)):
        mindistances.loc[j,i] = mindistance(testDimensions[i].loc[j],newDimensions[i])
```

```python
avgmindistances = np.empty(15,dtype = object)
for i in list(range(0,15)):
    avgmindistances[i] = np.mean(mindistances[i])
```

We created 1000 1-15D points and 100 1-15D test points. Calculated 100 test point's nearest neighbor distances and took mean of them. Visualized those means as a function of dimensionality.
From graph, we can say that as dimensionality goes higher, nearest neighbor distances gets bigger. It confirms the result of distance to origin versus dimensionality graph. So, as dimensionality goes higher, it becomes harder to create clusters and find outliers.

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(list(range(1,16)),
          avgmindistances,
          color = "black")
ax.set(title= "Nearest Neighbor Distances versus Dimensions",
      xlabel = "Dimensions",
      ylabel = "Average Min Distances")

plt.show()
```

Photo object is a numpy.ndarray object. It has shape 512x512x3. Here 512x512 corresponds to pixels and x3 corresponds to rgb channels.

```python
photo = mpimg.imread("photo.jpg")
print(np.ndim(photo))
print(np.size(photo))
print(np.shape(photo))
print(type(photo))
imgplot = plt.imshow(photo)
```

Extracted each channel seperately from original photo.

```python
photo_red = np.zeros((512,512,3), dtype="uint8")
photo_green = np.zeros((512,512,3), dtype="uint8")
photo_blue = np.zeros((512,512,3), dtype="uint8")
photo_red[:,:,0] = photo[:,:,0]
photo_green[:,:,1] = photo[:,:,1]
photo_blue[:,:,2] = photo[:,:,2]
```

```python
plt.imshow(photo_red)
```

```python
plt.imshow(photo_green)
```

```python
plt.imshow(photo_blue)
```

```python
photo_red_average = np.zeros((1,512,3), dtype="uint8")
for i in list(range(0,512)):
    photo_red_average[0,i,0] = np.mean(photo_red[:,i,0])
photo_green_average = np.zeros((1,512,3), dtype="uint8")
for i in list(range(0,512)):
    photo_green_average[0,i,1] = np.mean(photo_green[:,i,1])
photo_blue_average = np.zeros((1,512,3), dtype="uint8")
for i in list(range(0,512)):
    photo_blue_average[0,i,2] = np.mean(photo_blue[:,i,2])
```

Red, green and blue average distribution is calculated and visualized. This graph is interesting and I think it shows us the distribution of rgb colors in original photo.

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(photo_red_average[0,:,0],color = "red")
ax.plot(photo_green_average[0,:,1],color = "green")
ax.plot(photo_blue_average[0,:,2],color = "blue")
ax.set(title= "Average RGB",
      xlabel = "x coordinates",
      ylabel = "y coordinates")

plt.show()
```

Substracted one half of image from other half. It created an interesting new image.

```python
photo_red_half = np.zeros((512,512,3), dtype="uint8")
photo_red_half[:,0:256,0] = photo_red[:,0:256,0] - photo_red[:,256:512,0]
photo_red_half[:,256:512,0] = photo_red[:,256:512,0]
plt.imshow(photo_red_half)
```

```python
photo_green_half = np.zeros((512,512,3), dtype="uint8")
photo_green_half[:,0:256,1] = photo_green[:,0:256,1] - photo_green[:,256:512,1]
photo_green_half[:,256:512,1] = photo_green[:,256:512,1]
plt.imshow(photo_green_half)
```

```python
photo_blue_half = np.zeros((512,512,3), dtype="uint8")
photo_blue_half[:,0:256,2] = photo_blue[:,0:256,2] - photo_blue[:,256:512,2]
photo_blue_half[:,256:512,2] = photo_blue[:,256:512,2]
plt.imshow(photo_blue_half)
```

```python
photo_half = np.zeros((512,512,3), dtype="uint8")
photo_half[:,:,0] = photo_red_half[:,:,0]
photo_half[:,:,1] = photo_green_half[:,:,1]
photo_half[:,:,2] = photo_blue_half[:,:,2]
plt.imshow(photo_half)
```

Added some random noises to each channel and original image. It became interesting.

```python
noisered = np.random.uniform(low=0, high=0.1*218, size = (512,512,3))
noisegreen = np.random.uniform(low=0, high=0.1*226, size = (512,512,3))
noiseblue = np.random.uniform(low=0, high=0.1*239, size = (512,512,3))
```

```python
photo_red_noisy = np.zeros((512,512,3), dtype="uint8")
photo_red_noisy[:,:,0] = photo_red[:,:,0] + noisered[:,:,0]
plt.imshow(photo_red_noisy)
```

```python
photo_green_noisy = np.zeros((512,512,3), dtype="uint8")
photo_green_noisy[:,:,1] = photo_green[:,:,1] + noisegreen[:,:,1]
plt.imshow(photo_green_noisy)
```

```python
photo_blue_noisy = np.zeros((512,512,3), dtype="uint8")
photo_blue_noisy[:,:,2] = photo_blue[:,:,2] + noiseblue[:,:,2]
plt.imshow(photo_blue_noisy)
```

```python
photo_noisy = np.zeros((512,512,3), dtype="uint8")
photo_noisy[:,:,0] = photo_red_noisy[:,:,0]
photo_noisy[:,:,1] = photo_green_noisy[:,:,1]
photo_noisy[:,:,2] = photo_blue_noisy[:,:,2]
plt.imshow(photo_noisy)
```
